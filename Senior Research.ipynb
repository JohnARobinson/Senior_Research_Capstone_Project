```python
def gradientdec():
    #print(weights[0][0])
    #print(weights2[3][49])
    
    
    #total error change with respect to the output
    backprop()
    etotal_respect_to_out1 = 0
    etotal_respect_to_out2 = 0
    etotal_respect_to_out3 = 0
    etotal_respect_to_out4 = 0
    
    etotal_respect_to_out1 = outputlayer[0]-errortarget1
    etotal_respect_to_out2 = outputlayer[1]-errortarget2
    etotal_respect_to_out3 = outputlayer[2]-errortarget3
    etotal_respect_to_out4 = outputlayer[3]-errortarget4
    
    #print(etotal_respect_to_out1)
    #print(etotal_respect_to_out2)
    #print(etotal_respect_to_out3)
    #print(etotal_respect_to_out1)
    #print(errortarget4)
    #print(outputlayer)
    
    etotal_respect_to_out1 = outputlayer[0]-errortarget1
    
    
    #output of o_1 change with respect to its total net input
    output_respect_to_net1 = 0
    output_respect_to_net2 = 0
    output_respect_to_net3 = 0
    output_respect_to_net4 = 0
    
    output_respect_to_net1 = outputlayer[0]*(1-outputlayer[0])
    output_respect_to_net2 = outputlayer[1]*(1-outputlayer[1])
    output_respect_to_net3 = outputlayer[2]*(1-outputlayer[2])
    output_respect_to_net4 = outputlayer[3]*(1-outputlayer[3])
    
    global weights2
    global weights
    global hiddenlayer
    
    #total net input of o1 change with respect to weights
       
    for q in range(0,4):
        for w in range(0,50):
            if(q == 0):
                weights2[q][w] = weights2[q][w] - learnrate * (etotal_respect_to_out1 * output_respect_to_net1 * hiddenlayer[w])
            if(q == 1):
                weights2[q][w] = weights2[q][w] - learnrate * (etotal_respect_to_out2 * output_respect_to_net2 * hiddenlayer[w])
            if(q == 2):
                weights2[q][w] = weights2[q][w] - learnrate * (etotal_respect_to_out3 * output_respect_to_net3 * hiddenlayer[w])
            if(q == 3):
                weights2[q][w] = weights2[q][w] - learnrate * (etotal_respect_to_out4 * output_respect_to_net4 * hiddenlayer[w])
                
    inputreset()           
    length = flat()
    #print(length)
    
    etotal1_respect_to_h = (outputlayer[0]-errortarget1*outputlayer[0]*(1-outputlayer[0]))*weights2[0][0]
    etotal2_respect_to_h = (outputlayer[1]-errortarget2*outputlayer[1]*(1-outputlayer[1]))*weights2[0][0]
    etotal3_respect_to_h = (outputlayer[2]-errortarget3*outputlayer[2]*(1-outputlayer[2]))*weights2[0][0]
    etotal4_respect_to_h = (outputlayer[3]-errortarget3*outputlayer[3]*(1-outputlayer[3]))*weights2[0][0]
    etotal_respect_to_h =  etotal1_respect_to_h + etotal2_respect_to_h + etotal3_respect_to_h +  etotal4_respect_to_h
    
    for q in range(0,50):
        #etotal_respect_to_h = (((outputlayer[0]-errortarget1*outputlayer[0]*(1-outputlayer[1])*weights2[0])+((outputlayer[1]-errortarget2*outputlayer[1]*(1-outputlayer[1])*weights2[0])+((outputlayer[2]-errortarget3*outputlayer[2]*(1-outputlayer[2])*weights2[0])+((outputlayer[3]-errortarget4*outputlayer[3]*(1-outputlayer[3])*weights2[0]))
        output_h_respect_to_net = hiddenlayer[q]*(1-hiddenlayer[q])
        for w in range(0,length):
            #output_h_respect_to_net = hiddenlayer[q]*(1-hiddenlayer[q])
            weights[q][w] = weights[q][w] - learnrate * (etotal_respect_to_h * output_h_respect_to_net * inputlayer[w])
         
         #almost done
         #https://mattmazur.com/2015/03/17/a-step-by-step-backpropagation-example/
         #https://www.youtube.com/watch?v=72Ez3mz0QTI
```
