```python
def backprop():
    age1 = 18
    age2 = 30
    age3 = 61
    target = 0
    
    global etotal
    global errortarget1
    global errortarget2
    global errortarget3
    global errortarget4
    global correctpred
    global correctanswer

    

#finds which bracket the age falls under
    if(int(correctanswer) < age1):
        errortarget1 = 0.99
        errortarget2 = 0.01
        errortarget3 = 0.01
        errortarget4 = 0.01
        
        correctpred = 1
        
    if(int(correctanswer) >= age1 and int(correctanswer) < age2):
        errortarget1 = 0.01
        errortarget2 = 0.99
        errortarget3 = 0.01
        errortarget4 = 0.01
        
        correctpred = 2
        
    if(int(correctanswer) >= age2 and int(correctanswer) < age3):
        errortarget1 = 0.01
        errortarget2 = 0.01
        errortarget3 = 0.99
        errortarget4 = 0.01
        
        correctpred = 3
        
    if(int(correctanswer) >= age3):
        errortarget1 = 0.01
        errortarget2 = 0.01
        errortarget3 = 0.01
        errortarget4 = 0.99
        
        correctpred = 4
        
#error rate
    
    etotal1 = 0.5*(errortarget1-outputlayer[0])**2
    etotal2 = 0.5*(errortarget2-outputlayer[1])**2
    etotal3 = 0.5*(errortarget3-outputlayer[2])**2
    etotal4 = 0.5*(errortarget4-outputlayer[3])**2
    
    etotal = etotal1+etotal2+etotal3+etotal4
```
